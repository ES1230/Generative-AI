```python
import gradio as gr

def bmi(name, height, weight, diet):
  bmi_val = round(weight / ((height/100) **2),2)
  # 제곱 & 소수점 둘째 자리까지

  emotion = ""

  if bmi_val >23:
    emotion = "😥"
  else:
    emotion = "😄"

  if diet:
    diet = "다이어트 중"
  else:
    diet = "살 찌는 중"


  return (f"안녕하세요 {name}님 당신의 bmi는 {bmi_val}입니다.",emotion,diet)


# 사용자 정의 테마 설정
custom_theme = gr.themes.Default(
    # color: slate, gray, zinc, neutral, stone, red, orange, amber, yellow, lime, green, emerald, teal, cyan, sky, blue, indigo, violet, purple, fuchsia, pink ,rose

    # 강조색상 톤 설정
    primary_hue= gr.themes.colors.blue,
    # 선택된 요소나 활성화된 요소의 색상 톤 설정
    secondary_hue= gr.themes.colors.green,
    # 앱의 배경색, 텍스트 색상, 비활성화된 요소의 색상 톤 설정
    neutral_hue= gr.themes.colors.red,
    # 텍스트 크기 설정 ("sm", "md", "lg",)
    text_size="lg",
    # 요소 간의 간격 크기 설정 ( "sm", "md", "lg",)
    spacing_size="lg",
    # 요소의 모서리 둥글기 크기 설정 ("sm", "md", "lg",)
    radius_size="sm",
    # 기본 폰트 설정
    font=[gr.themes.GoogleFont("Mooli"), "Arial", "sans-serif"],
    # 고정폭 폰트 설정
    font_mono=[gr.themes.GoogleFont("Inconsolata"), "Arial", "sans-serif"],
)



interface = gr.Interface(
    fn=bmi,
    inputs=["text",gr.Slider(1,300,label="키(cm)"),gr.Slider(1,150,label="몸무게(kg)"),gr.Checkbox(label="다이어트 여부")], #컴포넌트로 받기 (슬라이더)
    outputs=["text","text","text"],
    flagging_options=["good","bad","error"],  #플래깅 위한 옵션
    examples= [["류은선",185,90,False],["이순신",180,60,True]],
    title = "BMI 계산기",
    description="문제가 있다면 Flag 버튼을 누르세요!",
    theme='gradio/seafoam'
    #theme는 Base, Default, Glass, Monochrome, Soft등이 있음  // theme=gr.themes.Base()
    #https://huggingface.co/spaces/gradio/theme-gallery  테마갤러리 : 다른사람이 만들어둔 테마 쓸 수 있음
)

interface.launch(debug=True)
```
