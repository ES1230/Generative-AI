```python
import numpy as np
import gradio as gr


def flip_text(x):  #text 뒤집기 앱
    return x[::-1]


def flip_image(x): #이미지 좌우반전 앱
    return np.fliplr(x)


with gr.Blocks() as demo:
    gr.Markdown("Flip text or image files using this demo.")
    with gr.Tab("Flip text"):  #탭선택으로 바꿀 수 있음
      text_input = gr.Textbox()
      text_output = gr.Textbox()
      text_button = gr.Button("Flip")
    with gr.Tab("Flip Image"):
      with gr.Row():
        image_input = gr.Image()
        image_output = gr.Image()
      image_button = gr.Button("Flip")

    with gr.Accordion("Open for More!"):
        gr.Markdown("Look at me...")

    text_button.click(flip_text, inputs=text_input, outputs=text_output)
    image_button.click(flip_image, inputs=image_input, outputs=image_output)

demo.launch(share=True)
```

```python
from transformers import pipeline
import gradio as gr

classifier = pipeline("sentiment-analysis")

def sentiment_analysis(text):
  result = classifier(text)
  return result[0]['label']

#Gradio interface
interface = gr.Interface(
    fn = sentiment_analysis,
    inputs = gr.Textbox(),
    outputs=gr.Textbox(),
    examples=["I love this","I hate this"]
)

interface.launch(share=True)
```

```python
## Zero-Shot Classification
# 모델이 학습 중에 본 적 없는 카테고리에 대한 데이터를 분류할 수 있게하는 기술 == 학습 데이터셋에 없는 새로운 카테고리의 데이터를 예측
# 실제 세계의 데이터는 항상 변하고 있기 때문에, 모델이 학습 중에 본 적 없는 새로운 카테고리의 데이터를 처리할 수 있도록 하는 것이 중요

from transformers import pipeline

classifier = pipeline("zero-shot-classification")
classifier(
    "This is a course about the Transformers library",
    candidate_labels=["education", "politics", "business"],  #후보군, 정치학/교육/사업중 하나다 라고 후보를 알려주는 것
)
```

```python
from transformers import pipeline # Hugging Face의 transformers 라이브러리에서 pipeline을 가져옴
import gradio as gr #gradio라이브러리 가져오기

# 네임드 엔티티 인식(NER) 파이프라인 로드
ner = pipeline("ner", grouped_entities=True) #엔티티 그룹 묶기

#텍스트에서 엔티티 추출하는 함수 정의
def extract_entities(text):
  result = ner(text)
  entities={} #딕셔너리 초기화

  for entity_group in result:
    entities[entity_group['word']] = [entity_group['entity_group'],entity_group['score']]

  return entities

#Gradio 인터페이스 정의
interface = gr.Interface(
  fn=extract_entities,
  inputs=gr.Textbox(label="Text for NER"),
  outputs=gr.Textbox(label="results")
)

#인터페이스 실행
interface.launch(share=True, debug=True)
```

```python
#질의응답
from transformers import pipeline
import gradio as gr

#Question-Answering 모델 로드
question_answerer = pipeline("question-answering")

#질문-응답 함수 정의
def def_question_answer(question,context):
  result = question_answerer(question=question, context=context)
  return result['answer']

#Gradio 인터페이스 정의
interface = gr.Interface(
    fn=def_question_answer,
    inputs=[
        gr.Textbox(lines=2,label="Question"),
        gr.Textbox(lines=4,label="Context")
    ],
    outputs=gr.Textbox(label="Answer"),
    examples=[
        ["where do I work?", "My name is ESeon and I work at Google in Osan."],
        ["What is my favorite color?","I have many favorite colors but I like blue the most."]
    ]

)
#인터페이스 실행
interface.launch(share=True)
```

```python
from transformers import pipeline
import gradio as gr

# ASR 파이프라인 생성
pipe = pipeline("automatic-speech-recognition", model="facebook/wav2vec2-base-960h")
# 텍스트 분류 파이프라인 생성
classifier = pipeline("text-classification")


# 음성 -> 텍스트로 변환하는 함수 정의
def speech_to_text(speech):
  text=pipe(speech)["text"]
  return text

# 텍스트 감정 분류 수행 함수 정의
def text_to_sentiment(text):
  sentiment_label = classifier(text)[0]['label']
  sentiment_score = round(classifier(text)[0]['score'] * 100,2)
  return f"{sentiment_label}, {sentiment_score }%"

#Gradio 블록 사용하여 인터페이스 정의
with gr.Blocks() as demo:
  audio_file = gr.Audio(type="filepath")
  text = gr.Textbox()   # 텍스트를 입력할 수 있는 상자 설정
  label=gr.Textbox()    # 감정 결과 표시할 상자

  with gr.Row():
    button1 = gr.Button("Recognize Speech") #버튼이름
    button1.click(speech_to_text, inputs=audio_file, outputs=text)

    button2 = gr.Button("Classify sentiment")
    button2.click(text_to_sentiment, inputs=text, outputs=label)


demo.launch(share=True)


```

```python
from transformers import pipeline
import gradio as gr

# 이미지 분류 pipeline 초기화
pipe = pipeline("image-classification", model="google/vit-base-patch16-224")

def classify_image(image):
    results = pipe(image)
    return {result['label']: result['score'] for result in results}

# Gradio 인터페이스 설정
iface = gr.Interface(
    fn=classify_image,
    inputs=gr.Image(type="pil", label="Input Image"),
    outputs=gr.Label(num_top_classes=3, label="Classification Results"),
    title = "Remove Background"
)

# 인터페이스 실행
iface.launch()


```
