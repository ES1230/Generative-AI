```python
import gradio as gr
import openai

openai.api_key = ""

def chat_with_gpt(user_input): #chatgpt에게 말거는 함수, 형식 정해져 있음
    user_message ={
        "role" : "user",
        "content" : user_input
    }

    res=openai.ChatCompletion.create(
        model="gpt-3.5-turbo",
        messages=[user_message]
    )

    gpt_response = res["choices"][0]["message"]["content"]
    return gpt_response

interface = gr.Interface(
    fn = chat_with_gpt,
    inputs="text",
    outputs="text"
)

interface.launch(share=True)
```

```python
import gradio as gr
import random

# message는 사용자의 입력을 나타내는 문자열
# history는 지금까지의 대화를 나타내는 리스트

def random_response(message,history): #message와 history가 가장 중요
  print(f"message : {message}")
  print(f"history : {history}")

  if len(history) % 2 == 0 :
    return f"홀수번째 메세지 {message}"
  else :
    return f"짝수번째 메세지 {message}"

gr.ChatInterface(random_response).launch(share=True, debug=True)
```

```python
import gradio as gr
import openai

openai.api_key = ""

def predict(message, history):
  # OpenAI Chatgpt API에 전달할 메세지 형식
  history_openai_format=[]

  # 기존의 대화 목록을 OpenAI ChatGPT가 이해할 수 있도록 변환
  for human, assistant in history :
    history_openai_format.append({"role":"user", "content":human}) #사용자의 메세지를 추가
    history_openai_format.append({"role":"assistant", "content":assistant}) #chatGPT의 응답 추가

  #가장 최근에 입력된 사용자의 메세지 추가
  history_openai_format.append({"role":"user", "content":message})

  #OPENAI ChatGPT API호출
  response = openai.ChatCompletion.create(
      model="gpt-3.5-turbo",  #사용할 모델 설정
      #보낼 메세지(대화 기록 전송)
      messages=[{"role":"system","content":"당신은 웨이트 트레이닝 전문가입니다. 웨이트 트레이닝 이외의 답변은 하지 마시오. 헬스 관련 질문은 이모티콘을 사용해서 자세히 친절하게 답변해줘"}
                , *history_openai_format],  # 앞에 *을 붙이면 unpacking
      temperature=0.8, #모델의 응답을 얼마나 다양하게 할건지 결정 (0~1 사이의 값) -> 0이면 일관된 답변, 1이면 창의적인 답변
      stream=True #응답을 스트리밍 형태로 받을 것인지 여부 (한번에 받을건지, 스트리밍으로 조금씩 받을건지)
  )

  #응답 출력 (디버깅 용도)
  print(response)

  partial_message = ""

  #응답 스트림 처리
  for chunk in response:
    #응답이 비어있지 않으면 메세지 부분 추가
    if len(chunk['choices'][0]['delta']) !=0:
        partial_message += chunk['choices'][0]['delta']['content']
        yield partial_message   #부분 메세지 반환 (generator 사용)
                                #yield 는 함수의 return 같은 것 / list는 공간을 계속 차지하지만 generator는 쓰고 버림

#gradio 인터페이스 설정
gr.ChatInterface(
    fn=predict,
    textbox=gr.Textbox(placeholder="말 걸어 주실래요?", container=False, scale=7)
).queue().launch(share=True, debug=True)

# queue: 대기열을 만들어 챗봇의 성능을 극대화함 // queue 60초 이상 걸리면 통신을 차단 , 메세지를 쪼개고 대기열 [메세지1, 메세지2, 메세지3]
# launch: 인터페이스 실행, share=True로 외부에서도 접근 가능하게 설정, debug=True로 디버깅 모드 활성화
```
