* LangChain은 대규모 언어모델(LLM)을 기반으로 애플리케이션을 구축하는 것을 도와주는 프레임워크   
LangChain은 Agents, Chains, Memory, Prompts 등 여러 모듈로 이루어져 있어 필요한 부분만 활용하거나 직접 커스터마이징할 수 있음   
파이썬 & 자바스크립트로 사용할 수 있음

* 대규모 언어 모델(LLM,Large Language Model)   
대규모 언어 모델은 다양한 자연어 처리(NLP)작업을 수행하는 딥러닝 모델   
LLM은 사람의 언어나 복잡한 데이터를 인식하고 해석할 수 있을 정도로 수많은 양의 데이터를 학습한 모델

* 자연어 처리(NLP, Natural Language Processing)   
자연어 처리(NLP)는 인공지능의 한 분야로, 컴퓨터가 사람의 언어를 이해하고 해석할 수 있도록 하는 기술


### LangChain 특징

**(1) 강력한 기능들이 모듈화된 구조**

LangChain은 Agents, Chains, Memory, Prompts 등 여러 모듈로 이루어져 있어 필요한 부분만 활용하거나 직접 커스터마이징할 수 있습니다.

1. Agents: 특정 태스크를 수행하는 에이전트로, LLM(대형언어모델)과 상호작용하며 작업을 분할하고 도구를 활용합니다. 예를 들어 Python REPL 에이전트, 웹 브라우저 에이전트, 계산기 에이전트 등이 있습니다.
2. Chains: 연속적인 단계로 이루어진 프로세스를 정의합니다. 질의응답, 요약, 번역 등의 chain이 제공되며 직접 커스터마이징도 가능합니다. LLM과 데이터, 도구를 엮어 복잡한 작업을 수행할 수 있습니다.
    - 예: 검색 -> OpenAI(LLM) -> 계산
3. Memory: 대화 기록, 데이터, 중간 결과 등을 저장하고 관리하는 메모리 시스템입니다. 벡터 데이터베이스, Redis, Pinecone 등 다양한 백엔드를 지원합니다.
4. Prompts: LLM에 입력할 프롬프트를 생성하고 관리합니다. 예를 들어 체이닝, 반복, 조건부 프롬프트 등의 기법을 활용할 수 있습니다. 효과적인 프롬프팅을 통해 LLM의 성능을 높일 수 있습니다.

이들 구성 요소를 활용하여 LangChain은 LLM 기반 애플리케이션의 기능(추론, Q&A, 요약) 등 다양한 작업에 유용하게 활용될 수 있습니다.

**(2) 간편한 통합**

다양한 LLM(대형언어모델), 데이터 소스, 도구들을 쉽게 통합할 수 있습니다. OpenAI, Anthropic, Hugging Face 등의 모델을 지원합니다.

**(3) LCEL을 활용한 손 쉬운 개발**

LCEL(LangChain Expression Language)을 통해 복잡한 워크플로우를 쉽게 작성하고, 실행할 수 있습니다. 병렬화, 배치 처리 등의 기능도 지원합니다.

체인이 복잡해질 수록 각 단계에서 어떤 일이 일어나는지 파악하는 것이 매우 중요해집니다. LCEL을 활용하면, 사용자에게 진행중인 작업을 알려주거나, 체인을 오류를 감지하는 것이 매우 쉬워집니다.

**(4) 활발한 커뮤니티**

많은 기여자들이 있고, 문서화와 예제 코드가 잘 갖추어져 있어 활용하기 좋습니다.

**요약**   
한 줄 정리하자면, 이미 잘 만들어진 LLM을 활용해서 나만의 애플리케이션을 빠르고 정확하게 만들 수 있는 것
